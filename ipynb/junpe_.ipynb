### 検討する

前回の記事

[将棋ウォーズのデータを解析する - 盗んだ統計で走り出す](http://tosh1ki.github.io/data-an/2014/10/31/wars-an/)

で使ったデータをそのまま流用する．データの詳細についてはそちらを参照してほしい．

```python
import pymongo
from collections import Counter
import re
from scipy.stats import poisson
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

connection = pymongo.Connection('localhost',27017)
db = connection.warskifu
col = db.kifu
```

MongoDBから件のアカウントが指した対局の棋譜を抜き出す．

```python
user = u'junpe_'
userlist = list(col.find({"$or":[{"user0":user},{"user1":user}]}))
len(userlist)
```

DBには74件登録されていた．件のアカウントの大会での成績は144勝27敗だったので，大会における件のプレーヤーが指した全対局の半分弱しか取得できていない．残りの対局は10秒指し以外のルールで指したか，あるいは大会期間の始めのほう（だいたい10月1〜5日）で指したと考えられる．


### 件のプレーヤーの対局のURL

```python
for j in userlist:
    print 'http://shogiwars.heroz.jp:3002/games/' + j[u'_id']
```

件のプレーヤーと2局以上対局したプレーヤーを列挙してみる．

```python
user_count = []
user_count.extend([_user[u'user0'] for _user in userlist])
user_count.extend([_user[u'user1'] for _user in userlist])

counter = Counter(user_count)

for u, count in counter.most_common():
    print u, count
```

やけにツツカナとの対局が多いような気がする．

ツツカナとの対戦成績を見てみる．

```python
winner = []
for i in range(len(userlist)):
    csa = userlist[i][u'csa']
    userdata = userlist[i]
    caption = ""
    
    win = len(re.findall(r'[\+|-]',csa))%2
    if win==1: ## 先手勝ち
        _winner = userdata[u'user0']
    else:
        _winner = userdata[u'user1']
        
    caption =  _winner + '; {0} vs. {1}\n'.format(userdata[u'user0'], userdata[u'user1'])
    if caption.find('tsutsukana') > 0:
        winner.append(_winner)
        print caption,

print 
counter = Counter(winner)

for u, count in counter.most_common():
    print u, count
```

対ツツカナ(`1_tsutsukana` or `2_tsutsukana`)で23勝12敗であった．

次に件のプレーヤーの持ち時間の使い方を見てみる

```python
u = []
for i in range(len(userlist)):
    csa = userlist[i][u'csa']
    userdata = userlist[i]
    caption = ""
    
    win = len(re.findall(r'[\+|-]',csa))%2
    if win==1: ## 先手勝ち
        caption = userdata[u'user0']
    else:
        caption = userdata[u'user1']
        
    caption +=  ';  {0} vs. {1}\n'.format(userdata[u'user0'], userdata[u'user1'])
    caption += userdata[u'_id']
    
    user0_time = []
    user1_time = []
    
    for j,c in enumerate(csa.split('\n')):
        if (j+1)%4==0:
            user1_time.append(int(c.replace('T', '')))
        elif (j+1)%2==0:
            user0_time.append(int(c.replace('T', '')))
    
    _u = []
    if userdata[u'user0']==user:
        _u = user0_time
    elif userdata[u'user1']==user:
        _u = user1_time
    u.extend(_u)

    plt.plot(_u)
    plt.ylim(0,10)
    plt.title(caption)
    plt.figure()
```

上記のグラフそれぞれは，n手目(横軸)を指すときに使った持ち時間(縦軸)を表している．ここで，対戦相手の時間は無視して件のプレーヤーの時間だけ見ていることに注意する．また，図のタイトルの1行目は左から勝者，先手，後手のプレーヤーのidを表し，2行目は棋譜が載っているページのURLの末尾の文字列を表す．

件のプレーヤーのグラフは一見普通に見える．**やけにリズミカルに指している**感はあるがソフト指しと判定するには証拠不十分といった気がする．

ちなみに「普通でない持ち時間のグラフ」というのはどういったものを指すのかというと，例えばツツカナ(`1_tsutsukana`)はほとんどの指し手を考慮時間3秒で指すので，グラフを描くと以下のように一直線のグラフになる．

```python
csa = userlist[5][u'csa']
print userlist[5][u'user1']

user_time = []
for j,c in enumerate(csa.split('\n')):
    if (j+1)%4==0:
        user_time.append(int(c.replace('T', '')))
        
plt.plot(user_time)
plt.figure()
```

ついでに件のプレーヤーの持ち時間の使い方をすべてまとめてヒストグラムにしてみる．
